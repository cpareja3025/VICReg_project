```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cvn_data = pd.read_csv("./cvn (1).csv")
cvn_data.head()
```

```python
epochs = np.array(cvn_data.iloc[:,0])
train_losses = np.array(cvn_data.iloc[:,1])
train_accs = np.array(cvn_data.iloc[:,2])
val_losses = np.array(cvn_data.iloc[:,3])
val_accs = np.array(cvn_data.iloc[:,4])
```

```python
plt.plot(epochs,train_losses, label = "Train")
plt.plot(epochs, val_losses, label = "Validation")
plt.title("Train and Validation Losses")
plt.legend()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.savefig("/Users/cpare/repos/VICReg_project/plots/CVN_losses.png")
```

```python
plt.plot(epochs, train_accs, label = "Train")
plt.plot(epochs, val_accs, label = "Validation")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Train and Validation Accuracy")
plt.legend()
plt.savefig("/Users/cpare/repos/VICReg_project/plots/CVN_accs.png")
```
